```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
sns.histplot(data=df,x='sepal_length',y='species')
```

```python
sns.histplot(data=df,x='petal_length',y='species')
```

```python
sns.histplot(data=df,x='sepal_width',y='species')
```

```python
sns.histplot(data=df,x='petal_width',y='species')
```

```python
df.groupby(['species']).mean()
###just for fun..........
```

```python
fig, axes=plt.subplots(2,2,figsize=[16,9])
sns.histplot(df['petal_width'],ax=axes[0,0])
sns.histplot(df['petal_length'],ax=axes[0,1])
sns.histplot(df['sepal_width'],ax=axes[1,0])
sns.histplot(df['sepal_length'],ax=axes[1,1])
```

```python
sns.boxplot(data=df,x="species",y="sepal_length")
```
