### Naive Bayes is a probabilistic classification algorithm based on Bayes' theorem. 
It assumes that the presence of a particular feature in a class is independent of the presence of other features

 Naive Bayes is a classification algorithm that calculates the probability of a class given the features and selects the most probable class.

‚óè Na√Øve Bayes Classifier can be used for Classification of categorical data.

```python
import numpy as np 
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df
```

```python
X=df.iloc[:,0:4]
Y=df.iloc[:,-1]
```

```python
X
```

```python
Y
```

## test and train split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,x_test,Y_train,y_test=train_test_split(X,Y,test_size=0.3,shuffle=True)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gb=GaussianNB()
```

```python
gb
```

```python
gb.fit(X_train,Y_train)
```

```python
y_pred=gb.predict(x_test)
```

```python
y_pred
```

## Sccores

```python
gb.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay,precision_score, recall_score, f1_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(f"Error rate is {1-accuracy_score(y_test, y_pred)}")
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
disp=ConfusionMatrixDisplay(confusion_matrix=cm)
```

```python
disp.plot()
```

```python
def get_confusion_matrix_values(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    return(cm[0][0], cm[0][1], cm[1][0], cm[1][1])

TP, FP, FN, TN = get_confusion_matrix_values(y_test, y_pred)
print("TP: ", TP)
print("FP: ", FP)
print("FN: ", FN)
print("TN: ", TN)
```

```python
print("The Accuracy is ", (TP+TN)/(TP+TN+FP+FN))
print("The precision is ", TP/(TP+FP))
print("The recall is ", TP/(TP+FN))
```
