## 1. Seaborn Library Basics


Seaborn is a Python data visualisation library based on matplotlib. It provides a
high-level interface for drawing attractive and informative statistical graphics.

1. Distribution Plots

    a. Dist-Plot
    
    b. Joint Plot
    
    d. Rug Plot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
```

```python
df
```

```python
sns.catplot(x='sex',y='age',hue='survived',data=df)
```

```python
plt.show()
```

```python
sns.catplot(x='sex',hue='survived',kind='count',data=df)
```

```python
plt.show()
```

```python
sns.displot(df['fare'])
```

```python
plt.show()
```

### kde=False: 
This parameter specifies whether to show a kernel density estimation (KDE) curve on top of the histogram. In this case, it is set to False, so the KDE curve will not be displayed.

```python
sns.displot(df["fare"],kde=False,bins=10)
plt.show()
```

```python
sns.jointplot(x='age',y='fare',data=df)
```

```python
plt.show()
```

```python
sns.histplot(y='fare',data=df)
```

```python
plt.show()
```

```python
#sns.pairplot(df, hue="fare")
```

```python
sns.scatterplot(x = 'fare', y = 'pclass', hue = 'survived', data = df)
```

```python
plt.show()
```

```python
sns.catplot(x='pclass', y='fare', data=df, kind='bar')
```

```python
plt.show()
```
